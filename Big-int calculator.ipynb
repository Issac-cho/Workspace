```python
def compare(list1, list2):          #list1<=list2 이면 True, list1>list2 이면 False
        if len(list1)<len(list2):
            return True             #list2의 자리수가 더 크기 때문에 True
        elif len(list1) == len(list2):
            i = 0
            while list1[i] == list2[i]:
                i+=1
                if i==len(list1):
                  return True       #list1=list2 이기 때문에 True
            if list1[i]<list2[i]:
                return True         #list1<list2 이므로 True
            else:                   #list1[i]>list2[i]
                return False
        else:                       #len(list1)>len(list2)
            return False

def reverselist(a):
        for i in range(len(a)//2):
            tmp = a[i]
            a[i] = a[len(a)-1 - i]
            a[len(a)-1 - i] = tmp
        return a

class BigInt :
    
    def __init__(self, init = 0, sign = True, base = 2) :
        self.vec_rep = []     # list storing coefficients (private instance variable)

        self.base = base

        # check type of sign
        if isinstance(sign, bool):
            self.sign = sign   
        else:
            raise TypeError("sign should be of bool type")
            
        # check type of base
        if isinstance(base, int) and base > 1:
            self.base = base
        else:
            raise ValueError("base should be of positive integer larger than 1")
                
        # check type of init                
        if isinstance(init, int) and init < base and init > -1:
            self.vec_rep = [init]
            
        elif isinstance(init, list) and [-1<init[i]<base for i in range(len(init))] == [True for i in range(len(init))]:
            self.vec_rep = init.copy()
            
        else: 
            raise ValueError("init should be a list of non-negative integers less than base")

    def length(self) :
        return max([n for n,c in enumerate(self.vec_rep) if c != 0.0]+[0])+1    
    
    def get_integer(self):
        if self.sign == True:
            return sum(self.vec_rep[i]*pow(self.base,i) for i in range(self.length()))
        else:
            return -sum(self.vec_rep[i]*pow(self.base,i) for i in range(self.length()))
    def __neg__(self) :
        c = [self.vec_rep[i] for i in range(self.length())]
        return BigInt(init = c, sign = not self.sign, base = self.base)

    def __floordiv__(self, divisor):
        if self.base != divisor.base:
            raise ValueError('Bases {} and {} are different!'.format(self.base,divisor.base))
        if divisor==0:
            raise ValueError('Divisor cannot be 0')
        for i in range(divisor.length()-1,-1,-1):
            if divisor.vec_rep[i] != 0:
              break
            if i==0:
              raise ValueError('Divisor cannot be 0')
            

        a, b, B = self, divisor, self.base
        a.vec_rep = reverselist(a.vec_rep)
        b.vec_rep = reverselist(b.vec_rep)
        while a.vec_rep[0]==0:
            a.vec_rep.pop(0)
        while b.vec_rep[0]==0:
            b.vec_rep.pop(0)
        k, l = self.length(), divisor.length()

        q=[]
        for i in range(k):
            q.append(0)

        r=[]
        for i in range(k):
            r.append(a.vec_rep[i])
            while r[0] == 0:
                    r.pop(0)
            if not compare(b.vec_rep, r):           # b>r 이면 q[i]=0 으로 유지하고 continue
                continue

            while compare(b.vec_rep, r):            # r>=b 이면 반복
                rev = 0
                if len(r) == len(b.vec_rep):
                    rev = 0
                else:
                    rev = -1

                carry = 0
                for j in range(len(r)-1,0,-1):           # b와 r은 자리수 차이가 0 아니면 1이므로
                    tmp = r[j] - b.vec_rep[j+rev] + carry
                    carry = tmp//B
                    r[j] = tmp % B
                
                if rev == 0:
                    r[0] = r[0] - b.vec_rep[0] + carry
                else:                               # rev == 1
                    r[0] = r[0] + carry
                while r[0]==0:
                    r.pop(0)
                    if len(r)==1:
                        break    
                q[i]+=1
        
        q=reverselist(q)
        r=reverselist(r)
        if self.sign != divisor.sign:
            q[0] += 1
        
        a.vec_rep = reverselist(a.vec_rep)
        b.vec_rep = reverselist(b.vec_rep)
        return BigInt(init=q, sign= self.sign and divisor.sign, base = self.base)
    
    def __mod__(self, divisor):
        if self.base != divisor.base:
            raise ValueError('Bases {} and {} are different!'.format(self.base,divisor.base))
        if divisor==0:
            raise ValueError('Divisor cannot be 0')
        for i in range(divisor.length()-1,-1,-1):
            if divisor.vec_rep[i] != 0:
              break
            if i==0:
              raise ValueError('Divisor cannot be 0')
            
        a, b, B = self, divisor, self.base
        a.vec_rep = reverselist(a.vec_rep)
        b.vec_rep = reverselist(b.vec_rep)
        while a.vec_rep[0]==0:
            a.vec_rep.pop(0)
        while b.vec_rep[0]==0:
            b.vec_rep.pop(0)
        k, l = self.length(), divisor.length()

        r=[]
        for i in range(k):
            r.append(a.vec_rep[i])
            while r[0] == 0:
                    r.pop(0)
            if not compare(b.vec_rep, r):           # b>r 이면 q[i]=0 으로 유지하고 continue
                continue

            while compare(b.vec_rep, r):            # r>=b 이면 반복
                rev = 0
                if len(r) == len(b.vec_rep):
                    rev = 0
                else:
                    rev = -1

                carry = 0
                for j in range(len(r)-1,0,-1):           # b와 r은 자리수 차이가 0 아니면 1이므로
                    tmp = r[j] - b.vec_rep[j+rev] + carry
                    carry = tmp//B
                    r[j] = tmp % B
                
                if rev == 0:
                    r[0] = r[0] - b.vec_rep[0] + carry
                else:                               # rev == 1
                    r[0] = r[0] + carry
                while r[0]==0:
                    r.pop(0)
                    if len(r)==1:
                        break    

        r=reverselist(r)
        a.vec_rep = reverselist(a.vec_rep)
        b.vec_rep = reverselist(b.vec_rep)
        if self.sign != divisor.sign:
            carry = 0
            c=[]
            for i in range(len(r)):
                tmp = b.vec_rep[i] - r[i] + carry
                carry = tmp//B
                c.append(tmp % B)
            for i in range(len(r), l):
                tmp = b.vec_rep[i] + carry
                carry = tmp//B
                c.append(tmp % B)
            r=c

        return BigInt(init=r , sign=divisor.sign , base=self.base )

def divmod(self, divisor):
        return self//divisor, self%divisor
```
